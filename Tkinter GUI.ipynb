```python
import tkinter as tk
from tkinter import*

root = tk.Tk()

root.title("Learning Radiobutton")
root.geometry("500x500")
root.configure(bg="Black")
root.attributes("-alpha",0.9)

v = tk.IntVar()
L1 = Label (root,text="Choose a Programming Language",justify=tk.LEFT,padx=2,bg="black",fg="red")
L1.pack()

v=IntVar()

B1 = Radiobutton(root, text="Python", padx = 20, variable = v, value = 1,bg="black",fg="red" )
B1.pack()

B2 = Radiobutton(root, text="Javascript", padx = 20, variable = v, value = 2 ,bg="black",fg="red")
B2.pack()

B3= Radiobutton(root, text="Oracle", padx = 20, variable = v, value = 3 ,bg="black",fg="red")
B3.pack()
    
root.mainloop()
```

```python

root = tk.Tk()

root.title("Learning Checkbutton")
root.geometry("500x500")
root.configure(bg="Black")

L1 = Label (root,text="Choose a Fruit Name",justify=tk.LEFT,padx=2,bg="black",fg="red")
L1.grid(row=0,column=1)

v1=IntVar()

B1 = Checkbutton(root, text="Apple",variable=v1,fg="red")
B1.grid(row=1,column=1)

v2=IntVar()

B2 = Checkbutton(root, text="Orange",variable=v2,fg="red")
B2.grid(row=2,column=1)

v3=IntVar()

B3 = Checkbutton(root, text="Orange",variable=v3,fg="red")
B3.grid(row=3,column=1)

v4=IntVar()

B3 = Checkbutton(root, text="Grapes",variable=v4,fg="red")
B3.grid(row=4,column=1)

E1 = Entry(root, bd=5,fg="Red")
E1.grid(row=5,column=1)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import*

win = tk.Tk()
win.title("Mikey")
win.geometry("500x500")

mylable = tk.Label(win,text="Guess The Number Ten",bg="black",fg="Red")
mylable.grid(row=0,column=1)

def click_me1():
    B1.configure(text="looser")
    mylable.configure(text="20")
    
def click_me2():
    B2.configure(text="winner")
    mylable.configure(text="10")
    mylable.configure(bg="blue")
    
def click_me3():
    B3.configure(text="looser")
    mylable.configure(text="30")
    
def click_me4():
    B4.configure(text="looser")
    mylable.configure(text="40")

B1 = Button(win,text="Guess 10",bg="black",fg="Red",command=click_me1 , font=())
B1.grid(row=1,column=1)

B2 = Button(win,text="Guess 10",bg="black",fg="Red",command=click_me2)
B2.grid(row=2,column=1)

B3 = Button(win,text="Guess 10",bg="black",fg="Red",command=click_me3)
B3.grid(row=3,column=1)

B4 = Button(win,text="Guess 10",bg="black",fg="Red",command=click_me4)
B4.grid(row=4,column=1)


    
win.mainloop()
```
